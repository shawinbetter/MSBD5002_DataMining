```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

#### Load data

```python
file_path = "../train_valid.csv"
df = pd.read_csv(file_path,header=None)
X,y = np.array(df.iloc[:,0:784]),np.array(df[784])
```

#### Distribution

```python
y_count = Counter(y)
plt.style.use(['science','ieee','std-colors'])
fig = plt.figure(figsize=[8,4],dpi=200)
fig.patch.set_facecolor('#FFFFFF')
plt.grid(color='grey', linestyle='-.', linewidth=0.2)
plt.bar(range(10),[y_count[float(i)] for i in range(10)])
plt.xticks(range(10));
```

### Select different class images to visualize

```python
plt.style.use(['science','ieee','std-colors'])
fig = plt.figure(figsize=[8,5],dpi=150)
for i in range(10):
    index = np.where(y == float(i))[0][0]
    plt.subplot(2,5,i+1)
    plt.imshow(X[index].reshape(28,28),cmap='gray', vmin=0, vmax=1)
    plt.title(f"groundtruth {i}")
```
