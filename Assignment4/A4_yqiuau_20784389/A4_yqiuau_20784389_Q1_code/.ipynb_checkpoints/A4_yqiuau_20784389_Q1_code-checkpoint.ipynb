```python
import numpy as np
np.random.seed(666)
import pandas as pd
import matplotlib.pyplot as plt
```

### Load data and label

```python
path = '../EM_Points.txt'
points = np.loadtxt(path)
data,label = points[:,0:2],points[:,2]
```

### Define EM Class

```python
class EM:
    
    def __init__(self,k):
        self.k = k
        self.centers = []
        
    def distance(self,x,y):
        """
        Coumpute l2 distance
        
        x,y : array points
        """
        return np.linalg.norm(x-y)
    
    def random_pick_k_center(self):
        """
        To Pick k points as clusters
        """
        index_of_points = np.random.choice(range(self.n),self.k) #index of data
        self.indexes = index_of_points #to record index
        
        for i in index_of_points: #self attribute
            self.centers.append(self.data[i,:])
            
    def cal_weight(self,row,col):
        """
        calculate the value of weight matrix based on given row & col
        
        """
        
        ## Following code is based on formula in slides, no need to explain
        nominator = 1/np.power(self.distance(self.data[row,:],self.centers[col]),2)
            
        denominator = 0
        
        for i in self.centers:
            denominator += 1/self.distance(self.data[row,:],i)
        
        return nominator / denominator
    
    def cal_sse(self,p = 1):
        """
        calculate the total sse based on current cluster
        
        p: parameter, defual to 1, which mean the weight of distance-sqaure is linear
        """
        w = np.power(self.weights,p)
        
        dis_mat = np.zeros((self.n,self.k))
        
        for row in range(self.n):
            for col in range(self.k):
                dis_mat[row,col] = np.power(self.distance(self.data[row,:],self.centers[col]),2) #cal distance
        
        return np.sum(np.multiply(w,dis_mat))
    
    def fit(self,data,eps=1e-4):
        """
        fit function to iteratively find clusters
        
        data: np array data
        eps: tolerance of falut
        """
        self.data = data
        
        #record data attributes
        n = data.shape[0]
        self.n = n
        self.features = data.shape[1]
        
        #create empty weight matrix. Note that the data type should float
        self.weights = np.random.rand(n,self.k)
        
        # pick k cluster
        self.random_pick_k_center()
        
        #termination criteria
        delta = float('inf')
        SSE = float('inf')
        
        counter = 0 #for report use only
        
        while delta > eps:
            
            # ********** Expectation Part*******************
            for row in range(self.n):
                for col in range(self.k):
                    self.weights[row,col] = self.cal_weight(row,col) #cal weight matrix
            
            if counter == 0: #in the first iteration, the weight of cluster points should be treated specifically
                for i in range(self.k):
                    self.weights[self.indexes[i],:] = 0
                    self.weights[self.indexes[i],i] = 1
            
            # ********** Maximization Part*******************
            for i in range(self.k):
                ## Following code follows the formula in slides
                
                c_new = np.array([0.0]*self.features)
            
                for feature in range(self.features):

                    c_new[feature] = np.sum(np.multiply(self.weights[:,i],self.data[:,feature]))/np.sum(self.weights[:,i])
                
                
                self.centers[i] = c_new
            
            # new sse
            new_SSE = self.cal_sse()
            #compute change
            delta = np.abs(SSE - new_SSE) if counter > 0 else new_SSE
            SSE = new_SSE
            
            counter += 1
            
            
            if counter <= 2:
                print("Center for Iteration {} is {}".format(counter,self.centers))
                print('SSE for Iteration {} is {}'.format(counter,new_SSE))   
                print('-------------------------------')
            
        print("Final Center are {}".format(self.centers))
        print("Final SSE is {}".format(SSE))
    
    def plot(self):
        """
        plot the result
        """
        df = pd.DataFrame(self.data)
        labels = []
        
        for row in range(self.n):
            label,min_dis = 0,float('inf')
            
            for center in range(self.k):
                dis = self.distance(self.data[row,:],self.centers[center])
                label = center if dis < min_dis else label #compute final label
                min_dis = dis if dis < min_dis else min_dis
                
            labels.append(label)
        
        df['Label'] = labels
        
        fig = plt.figure(figsize=(12,8),dpi=200)
        plt.scatter(df[0], df[1], c=df['Label'])
```

```python
em = EM(k=3)
em.fit(data,1e-4)
em.plot()
```
